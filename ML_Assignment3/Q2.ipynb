**Perform K-Means Clustering for the crime_std data and identify the number of clusters formed and draw inferences.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
x = pd.read_csv('crime_std.csv')
x

```

```python
x.drop('Unnamed: 0',inplace=True,axis=1)
x
```

```python
x.describe()
```

```python

x.info()
```

```python
# collect wss for every k value 
wss = []

# collect within cluster variations for k from 1 to 10
for k in range(1, 11):
    # create an object of kmeans
    kmeans_model = KMeans(n_clusters=k, random_state=12345)

    # fit the data
    kmeans_model.fit(x)

    # collect the within cluster variation
    wss.append(kmeans_model.inertia_)

print(wss)
```

```python
plt.plot(range(1,11),wss)
plt.xlabel("Clusters")
plt.ylabel('WSS')
plt.title('Elbow Method')
```

**Optimum cluster number = 4**

```python
x1 = KMeans(n_clusters=4)
x1.fit(x)
x1.labels_
```
