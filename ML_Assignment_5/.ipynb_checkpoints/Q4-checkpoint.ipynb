### Q4 Consider following datasets and do PCA on each datasets.
Datasets: wine_std.csv, framingham_std.csv, universities_std.csv, crime_std.csv, Grocery_std

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
x = pd.read_csv("crime_std.csv")
x
```

```python
pca = PCA(n_components=1)
pca.fit(x)
new = pca.transform(x)

df = pd.DataFrame(new)

for i in range(df.shape[1]):
    x[f"Principal Component {i}"] = df[i]
x
```
