```python
df.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
df1 = df.columns

def scaling():
    for col in df1:
        from sklearn.preprocessing import StandardScaler
        scaler = StandardScaler()
        df[col] = scaler.fit_transform(df[[col]])
        
```
