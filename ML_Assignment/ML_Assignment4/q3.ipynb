Perform Hierarchical Clustering for the Universities_std data and identify the number of clusters formed.

```python
import numpy as np
import pandas as pd
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('universities_std.csv')
df
```

Data Pre-processing

```python
df.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
df.rename({'SAT':'sat','Top10':'top','Accept':'accept','SFRatio':'ratio','Expenses':'expense','GradRate':'gradrate'},inplace=True,axis=1)
df
```

```python
# single linkage hierarchical clustering
m1 = hierarchy.linkage(df,'single')
res1 = hierarchy.dendrogram(m1)
```

```python
# complete linkage hierarchical clustering
m2 = hierarchy.linkage(df,'complete')
res2 = hierarchy.dendrogram(m2)
```

```python
# average linkage hierarchical clustering
m3 = hierarchy.linkage(df,'average')
res3 = hierarchy.dendrogram(m3)
```
