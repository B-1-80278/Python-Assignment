Perform DBSCAN Clustering for the Grocery_std data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN 
from itertools import product
from sklearn.metrics import silhouette_score
```

### Data Pre-processing

```python
# drop the unnnamed column
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.columns
```

**Optimize eps and min_values**

```python
# generate the trial combinations of eps and min_values
eps_values = np.arange(1,15)
min_values = np.arange(2,15)
parameters = product(eps_values, min_values)

for (ep,min_value) in parameters:
    # create DBSCAN model with current eps, min_value
    model = DBSCAN(eps=ep,min_samples=min_value)
    # train the model
    model.fit(df)
    # get the cluster count of model
    clusters = len(np.unique(model.labels_))
    # print(clusters)
    if clusters > 1:
        silhouette = silhouette_score(df,model.labels_)
        print(ep, min_value, clusters ,silhouette)
```

**Optimum values : eps = 7, min_samples=2**
