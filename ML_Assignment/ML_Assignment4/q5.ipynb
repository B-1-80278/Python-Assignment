# Book

```python
bk = pd.read_csv('/home/sad7_5407/Desktop/Sunbeam Material/ML/Assignment/book.csv')
bk
```

### EDA

```python
bk.info()
```

```python
bk.describe()
```

### Pre-processing

```python
# to suppress the warning given by apriori (requires boolean dataframe)
bk.replace(to_replace=[0,1],value=[False,Tr,inplace=True)
bk
```

## Association Rules Mining

```python
rules = apriori(bk, min_support=0.1,use_colnames=True)
rules
```
