```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('glass.csv')
df
```

```python
df.info()
```

```python
df.corr()
```

```python
df = df.drop(['RI','Si','K','Ca','Fe'],axis = 1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Type'] = label_encoder.fit_transform(df['Type']) 
```

```python
df['Type'].value_counts()
```

```python
X = df.drop('Type',axis=1)
Y = df['Type']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.33)
```

```python
def Decision_Tree():
    
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def Random_Forest():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=11)
    model.fit(x_train,y_train)
    return model
    
```

```python
def CatBoost():
    from catboost import CatBoostClassifier
    model = CatBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def XGBoost():
    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Type'] = label_encoder.fit_transform(df['Type']) 
```

```python
dt_model = Decision_Tree()
rf_model = Random_Forest()
catboost_model = CatBoost()
xg_model = XGBoost()
```

```python
def evaluation(model):
    y_pred = model.predict(x_test)

    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    print(f"Accuracy = {accuracy_score(y_test, y_pred)*100:.2f}%")
    print(f"Precision Score = {precision_score(y_test, y_pred,average = 'weighted')*100:.2f}%")
    print(f"Recall Score = {recall_score(y_test, y_pred, average = 'weighted')*100:.2f}%")
    print(f"F1 Score = {f1_score(y_test, y_pred, average = 'weighted')*100:.2f}%")
```

```python
evaluation(dt_model)
```

```python
evaluation(rf_model)
```

```python
evaluation(catboost_model)
```

```python
evaluation(xg_model)
```
