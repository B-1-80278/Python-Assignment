In the given dataset TVMarketing.csv you have given  sales based on 'TV' marketing budget.
Build a ANN model to predict 'Sales' using 'TV' as the predictor variable and evaluate the model.

Dataset : TVMarketing.csv


```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

### Pre-processing

```python
X = df['TV']
Y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y, train_size=0.9)
```

### Model Building

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
tf.keras.backend.clear_session()
```

```python
model = Sequential(
    [
        Dense(units=5, input_shape=[1])
    ]
)

# add an output layer
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',loss='mean_absolute_error')
```

```python
model.fit(x_train,y_train,epochs=1000)
```

```python
def evaluate_model(model):
    y_pred = model.predict(x_test)
    from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
    print(f'mae : {mean_absolute_error(y_test,y_pred)}')
    print(f'mse : {mean_squared_error(y_test,y_pred)}')
    print(f'rmse : {np.sqrt(mean_squared_error(y_test,y_pred))}')
    print(f'r2 scroe : {r2_score(y_test,y_pred)}')
```

```python
evaluate_model(model)
```
